```python
from sklearn.metrics import accuracy_score, classification_report

# Example test dataset (text and corresponding true labels)
test_texts = ["The movie was fantastic!", "I did not like the food", "It was an average experience"]
true_labels = ["5 stars", "1 star", "3 stars"]  # Replace with actual star ratings corresponding to your test data

# Use the classifier to predict sentiment for the test dataset
predictions = classifier(test_texts)

# Extract the predicted labels
predicted_labels = [pred['label'] for pred in predictions]

# Calculate accuracy
accuracy = accuracy_score(true_labels, predicted_labels)
print(f"Accuracy: {accuracy:.4f}")

# Pre-define the full set of possible labels for the report
all_possible_labels = ["1 star", "2 stars", "3 stars", "4 stars", "5 stars"]

# Generate a detailed classification report, ensuring all labels are included
print("Classification Report:")
print(classification_report(true_labels, predicted_labels, labels=all_possible_labels, target_names=all_possible_labels))
```

```python

from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
from sklearn.metrics import accuracy_score, classification_report

# Load the RoBERTa model fine-tuned for Twitter sentiment analysis
model_name = "cardiffnlp/twitter-roberta-base-sentiment"

# Load the model and tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Create the sentiment analysis pipeline
classifier = pipeline('sentiment-analysis', model=model, tokenizer=tokenizer)

# Example test dataset (tweets and corresponding true labels)
tweets = ["The movie was fantastic!", "I did not like the food", "It was an average experience"]


# True labels for the tweets .
true_labels = ["positive", "negative", "neutral"]  # Assuming you have 3 classes: positive, negative, neutral

# Use the classifier to predict sentiment for the tweets
predictions = classifier(tweets)

# Extract the predicted labels
predicted_labels = []
for pred in predictions:
    # Mapping model's output to match with true labels
    label = pred['label']
    if label == "LABEL_0":
        predicted_labels.append("negative")
    elif label == "LABEL_1":
        predicted_labels.append("neutral")
    elif label == "LABEL_2":
        predicted_labels.append("positive")

# Calculate accuracy
accuracy = accuracy_score(true_labels, predicted_labels)
print(f"Accuracy: {accuracy:.4f}")

# Generate a detailed classification report
print("Classification Report:")
print(classification_report(true_labels, predicted_labels, target_names=["negative", "neutral", "positive"]))
```

```python


#Now lets do the Twitter Sentiment Analysis.(Twitter US Airline Sentiment)
import pandas as pd
import re
from sklearn.model_selection import train_test_split
from google.colab import files
uploaded = files.upload()
file_name = list(uploaded.keys())[0]  # Get the name of the uploaded file
df = pd.read_csv(file_name)

# Display the first few rows of the DataFrame to confirm it loaded correctly
df.head()
#Now we do preprocessing to clean the text data.
def preprocess_tweet(tweet):
  tweet = re.sub(r'http\S+|www\S+|https\S+', '', tweet, flags=re.MULTILINE)  # Remove URLs
  tweet = re.sub(r'\@\w+|\#', '', tweet) # Remove @mentions and hashtags.
  tweet = re.sub(r'\W', ' ', tweet)  # Remove special characters and punctuation
  tweet = re.sub(r'\s+', ' ', tweet).strip().lower()  # Remove extra spaces and lowercase
#Now Apply the preprocessing function to the tweet coloumn.
df['cleaned_text'] = df['text'].apply(preprocess_tweet)
# Check the structure of your dataset
df[['text', 'cleaned_text', 'airline_sentiment']].head()

# Split the data into training and testing sets
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['airline_sentiment'])

# Ensure all text data is of type string and check for NaN or missing values
train_df['cleaned_text'] = train_df['cleaned_text'].astype(str)
test_df['cleaned_text'] = test_df['cleaned_text'].astype(str)

# Ensure there are no NaN values
train_df = train_df.dropna(subset=['cleaned_text', 'airline_sentiment'])
test_df = test_df.dropna(subset=['cleaned_text', 'airline_sentiment'])

# Convert text data to list of strings
train_texts = train_df['cleaned_text'].tolist()
test_texts = test_df['cleaned_text'].tolist()

# Verify that the lists contain only strings
assert all(isinstance(text, str) for text in train_texts), "Non-string value found in train_texts"
assert all(isinstance(text, str) for text in test_texts), "Non-string value found in test_texts"




```
